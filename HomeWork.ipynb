```python
from sympy import Symbol, diff, oo, Interval, is_increasing, is_decreasing
from sympy.solvers import solve
from sympy.plotting import plot
x = Symbol('x', real = True)
f = x**2 - 6*abs(x) + 8
f
```

```python
#1.определить корни
roots = solve(f,x)
print(f'{len(roots)} корня:')
roots
```

```python
#2. Найти интервалы, на которых функция возрастает

derivate = diff(f,x)

roots=solve(derivate,x)
if len(roots)==1:
    root=roots.pop()
#if derivate.subs(x,root + 0.001)>0:
if is_increasing(f, Interval(root,oo)):
    print(f'Function increase in {Interval(root,oo)}')
    
#3. Найти интервалы, на которых функция убывает

derivate = diff(f,x)
roots=solve(derivate,x)
if len(roots)==1:
    root=roots.pop()
#if derivate.subs(x,root - 0.001)<0:
if is_decreasing(f, Interval(oo,root)):
    print(f'Function decrease in {Interval(-oo,root)}')
else:
     print(f'Function decrease in {Interval(root,oo)}')
```

```python
#4. Построить график
plot(f)
```

```python
#5. Вычислить вершину
extremum = f.subs(x,root)
extremum
```

```python
#6. Определить промежутки, на котором f > 0
roots = solve(f,x)
root1=roots.pop()
root2=roots.pop()
if f.subs(x,root1-0.01)>0:
    print(f'Function > 0 in {Interval(-oo,root1)}')
else:
    print(f'Function < 0 in {Interval(-oo,root1)}')
middle=(root2+root1)/2
middle
```
